### 04_Apply, Map

```python
url = 'https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv'
df = pd.read_csv(url)
df.shape
```

```python
df.Income_Category
```

```python
dic = {
    'Unknown'        : 'N',
    'Less than $40K' : 'a',
    '$40K - $60K'    : 'b',
    '$60K - $80K'    : 'c',
    '$80K - $120K'   : 'd',
    '$120K +'        : 'e'   
}

df['newIncome'] = list(map(lambda x : dic[x], df.Income_Category))
a = df['newIncome']
a
```

```python
def new_category(x):
    if x == 'Less than $40K':
        return 'a'
    elif x == '$40K - $60K':
        return 'b'
    elif x == '$60K - $80K':
        return 'c'
    elif x == '$80K - $120K':
        return 'd'
    elif x == '$120K +':
        return 'e'
    else:
        return 'N'

df['newIncome'] = df.Income_Category.apply(new_category)
df['newIncome']
```

```python
df.Customer_Age.unique()
```

```python
def new_age(x):
    if x <= 0 and x <= 9:
        return 0
    elif x <= 10 and x <= 19:
        return 10
    elif x <= 20 and x <= 29:
        return 20
    elif x <= 30 and x <= 39:
        return 30
    elif x <= 40 and x <= 49:
        return 40
    elif x <= 50 and x <= 59:
        return 50
    elif x <= 60 and x <= 69:
        return 60
    elif x <= 70 and x <= 79:
        return 70
    
df['AgeState'] = df.Customer_Age.apply(new_age)
df.AgeState.head()
```

```python
df.AgeState.value_counts()
```

```python
def func(x):
    if 'Graduate' in x:
        return 1
    else:
        return 0
    
df['newEduLevel'] = df.Education_Level.apply(func)
df.newEduLevel.value_counts()
```

```python
def func(x):
    if x >= 4500:
        return 1
    else:
        return 0

df['newLimit'] = df.Credit_Limit.apply(func)
df.newLimit.value_counts()
```

```python
def func(df):
    if df.Marital_Status == 'Married' and df.Card_Category == 'Platinum':
        return 1
    else:
        return 0

df['newState'] = df.apply(func, axis=1)
df.newState.value_counts()
```

```python
def func(x):
    if x == 'M':
        return 'male'
    else:
        return 'female'
    
df['Gender'] = df.Gender.apply(func)
df.Gender.value_counts()
```

```python
url = 'https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv'
df = pd.read_csv(url)
df.head()
```

```python
df.info()
```

```python
df['(년-월-일:시)'].head()
```

```python
def change_date(x):
    import datetime
    hour = x.split(':')[1]
    date = x.split(":")[0]
    
    if hour =='24':
        hour ='00:00:00'
        
        FinalDate = pd.to_datetime(date +" "+hour) +datetime.timedelta(days=1)
        
    else:
        hour = hour +':00:00'
        FinalDate = pd.to_datetime(date +" "+hour)
    
    return FinalDate

df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(change_date)

Ans = df
Ans.head()
```

```python
df['dayName'] = df['(년-월-일:시)'].dt.day_name()
df.dayName.head()
```
